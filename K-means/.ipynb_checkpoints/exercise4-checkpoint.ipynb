## 实验代码

```python
sms = pd.read_csv('./testSet.txt', sep='\t', header=None)
# sms
data_set = sms.values
```

```python
def selectInitCenter(data_set, k):
    """
    随机选择k个数据点作为簇的中心
    :param data_set: 待聚类的数据集
    :param k: 簇的个数
    :return: 以簇编号为key,中心点坐标为value的字典
    """
    random_index = random.sample(range(len(data_set)), k=k) # 随机选取k个点作为簇中心
    center = {}
    for i in range(k):
        center[i] = data_set[random_index[i]] # 以簇编号为key，中心点坐标为value构建字典
    return center
```

```python
def check(old_center,new_center):
    """
    判断迭代前后所有簇中心是否相等
    :param old_center: 迭代前所有簇中心，以字典形式表示，key为簇编号，value为中心点坐标
    :param new_center: 迭代后所有簇中心，同上
    :return: True->相等；False->不相等
    """
    old_center_list = np.array(list(old_center.values())) # 迭代前簇中心矩阵
    new_center_list = np.array(list(new_center.values())) # 迭代后簇中心矩阵
    reduce = np.abs(old_center_list-new_center_list) # 矩阵对应位置差的绝对值
    all_sum = np.sum(np.sum(reduce)) # 矩阵所有元素求和
    return all_sum <= 1e-9 # 浮点误差
```

```python
def calCenter(data_set, k, k_center):
    """
    迭代计算新的中心
    :param data_set: 数据集
    :param k: 簇的个数
    :param k_center: 迭代前的簇中心，以字典形式表示，key为簇编号，value为中心点坐标
    :return: 迭代后的中心
    """
    distance = []
    # 计算所有点与簇中心的距离组合成矩阵，Aij 表示第j个点与第i个簇中心的距离
    for i,center in enumerate(k_center.values()):
        dis = np.sum((data_set-center)**2,axis=1) # 计算每个点与簇中心的距离
        distance.append(list(dis))
    distance_sort = np.argmin(distance,axis=0) 
    new_center = {}
    for key in k_center.keys():
        filter_set = data_set[distance_sort==key,:] # 过滤出属于第i个簇的数据点
        new_center[key] = np.sum(filter_set,axis=0)/(len(filter_set)+1e-9) # 计算簇中心
    return new_center 
```

```python
def calFinalCenter(data_set, k, wich):
    """
    计算最终的中心点
    :param data_set: 数据集
    :param k: 簇的个数
    :return: 最终的中心点
    """
    if wich == 1: 
        old_center = selectInitCenter(data_set, k) # k-means
    else:
        old_center = selectInitCenter2(data_set, k) # k-means++
    new_center = calCenter(data_set, k, old_center)# 计算新的簇中心
    # 反复迭代，直至新的簇中心与旧簇中心相等
    while not check(old_center,new_center):
        old_center = new_center
        new_center = calCenter(data_set, k, new_center)
    return new_center
```

```python
def kmeans(data_set, k, wich=1):
    """
    k-means聚类
    :param data_set: 数据集
    :param k: 簇的个数
    :return: 
        final_center: k个簇中心坐标，以字典形式表示，key为簇编号，value为中心点坐标
        final_set: 簇样本点，以字典形式表示，key为簇编号，value为属于该簇的样本点集合
    """
    final_center = calFinalCenter(data_set, k, wich) # 计算最终的中心点
    distance = []
    for i,center in enumerate(final_center.values()):
        dis = np.sum((data_set-center)**2,axis=1)
        distance.append(list(dis))
    distance_sort = np.argmin(distance,axis=0)
    final_set = {}
    for key in final_center.keys():
        final_set[key] = data_set[distance_sort==key,:]
    return final_center, final_set
```

```python
def showCluster(final_center, final_set, K):
    """
    对聚类结果进行平面可视化，样本的维度需要为2维
    """
    mark = ['or', 'ob', 'og', 'ok', '^r', '+r', 'dr', '<r', 'pr'] # 图形种类
    # 检查
    # 1. 样本维度是否为2维
    # 2. 簇的数量是否超过可表示的图形种类
    if final_set[0].shape[1] != 2 or len(final_center) > len(mark):
        return
    for key in final_set.keys():
        plt.scatter(final_set[key][:,0], final_set[key][:,1], label='cluster{}'.format(key))
    plt.xlabel("x-coordinate")
    plt.ylabel("y-coordinate")
    plt.title("$K$ = {}".format(K))
    plt.legend()
    plt.savefig("cluster.svg", dpi=300,format="svg") # 保留图片
```

```python
def showClusterWithCenter(final_center, final_set, wich):
    """
    聚类结果以及中心点一起画出
    """
    mark = ['or', 'ob', 'og', 'ok', '^r', '+r', 'dr', '<r', 'pr'] # 图形种类
    for key in final_set.keys():
        plt.scatter(final_set[key][:,0], final_set[key][:,1], label='cluster{}'.format(key))
    center = np.array(list(final_center.values()))
    K = len(center)
    plt.scatter(center[:,0],center[:,1],marker='*', c='k', label='center')
    plt.xlabel("x-coordinate")
    plt.ylabel("y-coordinate")
    plt.title("$K$ = {}".format(K))
    plt.legend()
    if wich == 1:
        plt.savefig("clusterWithCenter.svg", dpi=300,format="svg") # 保留图片
```

## k-means++

```python
def selectInitCenter2(data_set, k):
    """
    从数据集中返回 k 个对象可作为质心
    """
    cluster_centers = {0:random.choice(data_set)} # 随机选取初始点
    for i in range(1, k):
        cluster_matrix = np.array(list(cluster_centers.values()))
        distance_list = []
        # 求每个点到所有簇中心点的距离的最小值Si
        for point in data_set:
            distance = np.min(np.sum((cluster_matrix-point)**2, axis=1))
            distance_list.append(distance)
        # 取Si最大值作为下一个质心
        cluster_centers[i] = data_set[np.argmax(distance_list)]
    return cluster_centers
```

## Silhouette coefficient

```python
def calAi(point, cluster, eps):
    return np.sum(np.sum((point-cluster)**2))/len(cluster-eps)
```

```python
def calBi(point, cluster_index, final_set):
    bi_list = []
    for key in final_set.keys():
        if cluster_index == key:
            continue
        bi_list.append(calAi(point,final_set[key], 0))
    return np.min(bi_list)
```

```python
def calSilhouette(final_set):
    count = 0
    S_i = 0.0
    for key in final_set.keys():
        cluster = final_set[key]
        count += len(cluster)
        for point in cluster:
            a_i = calAi(point,cluster,1)
            b_i = calBi(point,key,final_set)
            s_i = (b_i-a_i)/max(b_i,a_i)
            S_i += s_i
    return S_i/count
```

```python
def adjustKWitSilhouette(data_set, k_lower_bound, k_upper_bound):
    k_list = [i for i in range(k_lower_bound,k_upper_bound+1)]
    sil_list = []
    for k in k_list:
        final_center, final_set = kmeans(data_set, k)
        sil = calSilhouette(final_set)
        sil_list.append(round(sil,3)) # 保留3位小数
    return k_list, sil_list
```

## SSE

```python
def calSSE(final_center, final_set):
    sse = 0.0
    for key in final_center.keys():
        # 利用广播机制，计算每个簇中点与其中心点距离的平方
        distance = np.sum((final_set[key]-final_center[key])**2,axis=1)
        # 对距离平方求和，得到单个簇的误差平方和
        distance = np.sum(distance)
        # 计算累计误差平方和
        sse += distance
    return sse
```

```python
def adjustKWithSSE(data_set, k_lower_bound, k_upper_bound):
    """
    计算不同k值聚类结果下的误差平方和(SSE)
    :param data_set: 待聚类的数据集
    :param k_lower_bound: k值的下界
    :param k_upper_bound: k值的上界
    :return: k值列表，误差平方和列表。前者的值与后者的值一一对应
    """
    k_list = [i for i in range(k_lower_bound,k_upper_bound+1)]
    sse_list = []
    for k in k_list:
        final_center, final_set = kmeans(data_set, k)
        sse = calSSE(final_center, final_set)
        sse_list.append(sse)
    return k_list, sse_list
```

## k-medoids

k-medoids算法步骤：
1.任意选取k个初始中心点medoids；
2.按照与medoids最近的原则，将剩余点分配到当前最佳的medoids代表的类中;
3.在每一类中，计算每个样本点与其他点的距离之和，选取距离之和最小的点作为新的medoids;
4.重复2-3的过程，直到所有的medoids点不再发生变化，或已达到设定的最大迭代次数;
k-medoids算法选取簇中心点的准则函数是:当前簇中所有其他点到该中心点的距离之和最小，所以需要遍历簇中所有点；

```python
def calMinPoint(filter_set):
    """
    计算簇中与其他点距离之和最小的点
    """
    distance_sum_list = []
    for point in filter_set:
        distance_sum = np.sum(np.sum((point-filter_set)**2))
        distance_sum_list.append(distance_sum)
    return filter_set[np.argmin(distance_sum_list)]
```

```python
def calCenter2(data_set, k, k_center):
    """
    迭代计算新的中心
    :param data_set: 数据集
    :param k: 簇的个数
    :param k_center: 迭代前的簇中心，以字典形式表示，key为簇编号，value为中心点坐标
    :return: 迭代后的中心
    """
    distance = []
    # 计算所有点与簇中心的距离组合成矩阵，Aij 表示第j个点与第i个簇中心的距离
    for i,center in enumerate(k_center.values()):
        dis = np.sum((data_set-center)**2,axis=1) # 计算每个点与簇中心的距离
        distance.append(list(dis))
    distance_sort = np.argmin(distance,axis=0) 
    new_center = {}
    for key in k_center.keys():
        filter_set = data_set[distance_sort==key,:] # 过滤出属于第i个簇的数据点
        new_center[key] = calMinPoint(filter_set) # 计算簇中心
    return new_center
```

```python
def calFinalCenter2(data_set, k, wich):
    """
    计算最终的中心点
    :param data_set: 数据集
    :param k: 簇的个数
    :return: 最终的中心点
    """
    if wich == 1: 
        old_center = selectInitCenter(data_set, k) # k-means初始化质心方法
    else:
        old_center = selectInitCenter2(data_set, k) # k-means++初始化质心方法
    new_center = calCenter2(data_set, k, old_center) # 更新质心
    # 反复迭代，直至新的簇中心与旧簇中心相等
    while not check(old_center,new_center):
        old_center = new_center
        new_center = calCenter(data_set, k, new_center)
    return new_center
```

```python
def kmedoids(data_set, k, wich=2):
    """
    k-medoids聚类
    :param data_set: 数据集
    :param k: 簇的个数
    :return: 
        final_center: k个簇中心坐标，以字典形式表示，key为簇编号，value为中心点坐标
        final_set: 簇样本点，以字典形式表示，key为簇编号，value为属于该簇的样本点集合
    """
    final_center = calFinalCenter2(data_set, k, wich) # 计算最终的中心点
    distance = []
    for i,center in enumerate(final_center.values()):
        dis = np.sum((data_set-center)**2,axis=1)
        distance.append(list(dis))
    distance_sort = np.argmin(distance,axis=0)
    final_set = {}
    for key in final_center.keys():
        final_set[key] = data_set[distance_sort==key,:]
    return final_center, final_set
```

## k-means、k-means++、k-medoids 轮廓系数对比

```python
def calSilhouetteMutiTimes2(data_set,k, times, method:str):
    time_list = [i+1 for i in range(times)]
    sil_list = []
    for time in time_list:
        if method == 'k-means':
            final_center, final_set = kmeans(data_set, k, 1)
        elif method == 'k-means++':
            final_center, final_set = kmeans(data_set, k, 2)
        else:
            final_center, final_set = kmedoids(data_set, k, 1)
        sil_list.append(round(calSilhouette(final_set),3))
#         sil_list.append(round(calSSE(final_center, final_set),3))
    return time_list, sil_list
```
